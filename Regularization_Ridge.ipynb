```python
# Importing libraries
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
data = pd.read_csv('./prostate_cancer.txt')
data
```

```python
# Data preprocessing
data.info()
```

```python
sns.heatmap(data = data.corr(), annot= True, cmap='coolwarm')
```

```python
# Feature extraction
x = data[['lcavol', 'lcp', 'svi']]
y = data['lpsa']
print(x.shape)
print(y.shape)
```

```python
# Splitting the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.11, random_state=100)

# Creating a model
model = Ridge(alpha=1)

# Training the model
model.fit(x_train, y_train)

# Predicting the model
y_pred = model.predict(x_test)

# MAE, MSE, R2_score metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"R2 : {r2}")
```
